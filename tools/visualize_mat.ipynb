```python
from PIL import Image
from matplotlib import pyplot
import numpy as np
import torch
import torchvision

# img = Image.new("RGB",size=(224,224))
# pyplot.imshow(img)
# pyplot.show()
```

```python
with open("mat.txt","r") as fin:
    data = fin.read().split("\n")[:-1]
    vals = []
    for s in data:
        vals.append(float(s))
print(vals[:10])
```

```python
vals = np.array(vals)
print(vals[:10])
```

```python
vals = np.reshape(vals,(3,224,224))
vals = torch.tensor(vals)
print(vals.shape)
```

```python
inv_transform = torchvision.transforms.Compose([torchvision.transforms.Normalize(mean=[0., 0., 0.],
                                                                                     std=[1/0.229, 1/0.224, 1/0.225]),
                                                    torchvision.transforms.Normalize(mean=[-0.485, -0.456, -0.406],
                                                                                     std=[1., 1., 1.]),
                                                    torchvision.transforms.ToPILImage()
                                                    ])
```

```python
img = inv_transform(vals)
print(np.array(img).max(),np.array(img).min())
pyplot.imshow(img)
pyplot.show()
```
